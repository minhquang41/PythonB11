```python
s = input("Nhap 3 so (cach nhau boi dau cach): ").split()
a = float(s[0])
b = float(s[1])
c = float(s[2])
print(f"a = {a}, b = {b}, c = {c}")
delta = b*b - 4*a*c

if a == 0:
  if b == 0:
    if c == 0:
      print("Phuong trinh vo so nghiem")
    else:
      print("Phuong trinh vo nghiem")
  else:
    print("Phuong trinh co 1 nghiem x = ", -c/b)
else:
  if delta < 0:
    print("Phương trinh vo nghiem")
  elif delta == 0:
    print("Phương trinh co nghiem kep x = ", -b/(2*a))
  elif delta > 0:
    print("Phương trinh co 2 nghiem phan biet")
    print("x1 = ", (-b + delta**0.5)/(2*a))
    print("x2 = ", (-b - delta**0.5)/(2*a))
```
